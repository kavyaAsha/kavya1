# normalization

```python
from sklearn.preprocessing import StandardScaler
# finding the mean and std deviation of numerical columns
scaler = StandardScaler().fit(data[['JAN', 'FEB']]) 
# scaling columns to a common range
data_normalized = scaler.transform(data[['JAN', 'FEB']])# Scaled data 
#note that output is a 2-Dimensional array of normalized data
data_normalized
```

```python
#updating the numerical(normalized) data into the dataframe
data_normalized=pd.DataFrame(data_normalized,columns=['JAN', 'FEB'])
data_normalized=data_normalized.join(data[data.columns.drop(['JAN', 'FEB'])])
data_normalized
```

```python
plt.boxplot(data['JAN'])
```

```python
for i in data['JAN']:
    q1=data['JAN'].quantile(0.25)
    q3=data['JAN'].quantile(0.75)
    iqr=q3-q1
    lowertail=q1-1.5*iqr 
    uppertail=q3+1.5*iqr
    if i>uppertail or i<lowertail:
      data['JAN']=data['JAN'].replace(i,np.mean(data['JAN']))
plt.boxplot(data['JAN'])
plt.title("box plot ")    
plt.show()    
```
